# 1.Introduction to DDL (Data Definition Language)

Data Definition Language (DL) is a crucial sub-language of SQL used to define, modify, and manage the structural blueprint of a database. Unlike DML (Data Manipulation Language), which handles the actual rows of data, DDL commands operate directly on database objects such as tables, databases, schemas, indexes, and views.

Core Characteristics of DDL

Schema Level: It alters the structure or layout of the database, not the contents.

Auto-Commit: In most relational database management systems (RDBMS) like MySQL and Oracle, DDL operations are auto-committed. This means once you execute a DDL command, the changes are saved permanently to the disk immediately, and you cannot use ROLLBACK to undo them.


**There are four fundamental DDL commands used daily by database administrators and developers:**

 **CREATE, ALTER, DROP, and TRUNCATE.**

 




# A. CREATE

Used to build entirely new database components from scratch.


```python
CREATE DATABASE CollegeDB;
```

Creating a Table (With Constraints):

When creating a table, you must declare the column names, their exact data types, and any structural rules (constraints) they must follow.

```python
CREATE TABLE Students (
    StudentID INT PRIMARY KEY,                  -- Unique identifier, cannot be NULL
    FirstName VARCHAR(50) NOT NULL,             -- Text field that cannot be left blank
    LastName VARCHAR(50), 
    Age INT CHECK (Age >= 17),                  -- Validates data entry rules
    Email VARCHAR(100) UNIQUE,                  -- Ensures no duplicate emails exist
    EnrollmentDate DATE DEFAULT CURRENT_DATE    -- Automatically assigns today's date if omitted
);
```

# B. ALTER

Used to modify the structure of an existing table without wiping out the data currently stored inside it.



```python
 1. Adding a Column:
 
ALTER TABLE Students 
ADD PhoneNumber VARCHAR(15);
```

```python
2. Dropping (Removing) a Column:

ALTER TABLE Students 
DROP COLUMN LastName;
```

```python
3. Modifying a Column Data Type or Constraint:

ALTER TABLE Students 
MODIFY COLUMN FirstName VARCHAR(100) NOT NULL;
```

```python
4. Renaming a Column:

ALTER TABLE Students 
RENAME COLUMN Email TO StudentEmail;
```

# C. DROP
Completely destroys a database object. Running a DROP command deletes the entire structure and every single row of data inside it permanently.

```python
-- Deletes the entire table structure and data
DROP TABLE Students;

-- Deletes the whole database and everything inside it
DROP DATABASE CollegeDB;
```

# D. TRUNCATE
Often confused with DELETE, TRUNCATE is a DDL command used to wipe out all records from a table instantly while leaving the table structure (columns, data types, constraints) perfectly intact.

```python
TRUNCATE TABLE Students;
```

How it works under the hood: Instead of scanning and deleting rows one by one, TRUNCATE drops the table storage allocation and recreates an empty table shell. This makes it incredibly fast on large datasets.

# 2 Introduction to DML (Data Manipulation Language)

**Data Manipulation Language (DML)** is the sub-language of SQL responsible for managing, altering, and interacting with the actual data stored inside database tables. While DDL (Data Definition Language) sets up the empty structure and columns, DML is used to populate, modify, and clean up the records within that structure.

Core Characteristics of DML

Row Level: DML commands act directly on rows (records) within a table.

Non-Structural: Executing a DML command does not change the table's columns, data types, or constraints.

Transactional (Not Auto-Committed): Unlike DDL, DML operations are typically not permanently saved to the disk automatically. They can be reviewed, safely undone (ROLLBACK), or saved permanently (COMMIT) using Transaction Control Language (TCL).

There are three primary commands that make up the backbone of DML: INSERT, UPDATE, and DELETE.

# A. INSERT

The INSERT statement adds new rows of data into an existing table

**1. Inserting Values for All Columns**

If you are adding data for every single column in the exact order the table was created, you do not need to specify the column names.

```python
INSERT INTO Students 
VALUES (101, 'Swapnil', 'Rathore', 'swapnil@example.com', 21);
```

**2. Inserting Values for Specific Columns (Recommended)**

Specifying the columns explicitly is safer. It ensures your queries won't break if someone alters the table layout later by adding a new column. Any unmentioned columns will either receive their DEFAULT value or a NULL value.

```python
INSERT INTO Students (StudentID, FirstName, Email) 
VALUES (102, 'Amit', 'amit@example.com');
```

**3. Bulk Insert (Multi-row Insert)**

You can insert multiple records at once by separating the value sets with commas

```python
INSERT INTO Students (StudentID, FirstName, Email, Age) 
VALUES 
(103, 'Neha', 'neha@example.com', 22),
(104, 'Rohan', 'rohan@example.com', 20),
(105, 'Pooja', 'pooja@example.com', 23);
```

# B. UPDATE

The UPDATE statement modifies existing records within a table.

⚠️ CRITICAL WARNING: Always use a WHERE clause with an UPDATE statement. If you omit the WHERE clause, every single row in the table will be overwritten with the new values.

```python
--1. Updating a Single Column for a Specific Row

UPDATE Students 
SET Age = 22 
WHERE StudentID = 102;
```

```python
/*2. Updating Multiple Columns Simultanously
You can update multiple attributes at once by separating assignments with a comma.

*/

UPDATE Students 
SET Email = 'swapnil_new@example.com', Age = 22
WHERE StudentID = 101;
```

```python
/*3. Conditional Bulk Updates
You can update a group of records matching a broader condition.
*/

-- Giving a graceful increment to a status flag or tracking attribute
UPDATE Products
SET Status = 'Discontinued'
WHERE StockCount = 0;
```

# C. DELETE

The DELETE statement removes specific rows from a table based on a condition.

⚠️ CRITICAL WARNING: Just like UPDATE, omitting the WHERE clause in a DELETE statement will wipe out all rows from your table, leaving you with an empty table structure.

```python
--1. Deleting a Specific Record

DELETE FROM Students 
WHERE StudentID = 105;
```

```python
--2. Deleting a Group of Records
SQL
DELETE FROM Students 
WHERE Age < 18;
```

**3. Data Integrity & Constraints During DML Operations**

When executing DML commands, the database engine enforces the rules established by DDL constraints. If a DML command breaks a rule, the entire operation fails and throws an error.

Primary Key Violation: Trying to INSERT a StudentID that already exists will fail.

Foreign Key Violation: If an Orders table points to a CustomerID in a Customers table, trying to INSERT an order for a non-existent customer ID will fail.

Not Null Violation: Trying to insert a record without providing a value for a column marked NOT NULL will fail.

Check Constraint Violation: Attempting to UPDATE a student's age to 15 when a check constraint states CHECK (Age >= 17) will be rejected.

**4. DML Control via Transactions (TCL)**

Because DML operations alter valuable data, they run inside Transactions. A transaction is a unit of work that must succeed entirely or fail entirely (known as the Atomicity property in ACID).

```python
--Step 1: Begin the transaction block
START TRANSACTION;

-- Step 2: Perform DML changes
UPDATE Accounts SET Balance = Balance - 1000 WHERE AccountID = 501;
UPDATE Accounts SET Balance = Balance + 1000 WHERE AccountID = 502;

-- Step 3: Check condition. If something went wrong, discard changes:
-- ROLLBACK;

-- Step 4: If everything is successful, save permanently:
COMMIT;
```

# 3. Introduction to DCL (Data Control Language)

**Data Control Language (DCL)** is the sub-language of SQL that acts as the security system for your database. While DDL defines the tables and DML/DQL manipulate the data, DCL manages access control and security permissions.

When a database is used by multiple applications, developers, and analysts, you cannot allow everyone to have full access to everything. DCL ensures data confidentiality and integrity by restricting who can see or modify specific database objects.

 ***Core Components: Privileges and Users***

Before writing DCL commands, it helps to understand the two main concepts it manages:

Privileges (Permissions): The right to execute specific SQL commands. Examples include:

**SELECT (Read access)**

**INSERT (Write access)**

**UPDATE (Modify access)**

**DELETE (Remove access)**

**EXECUTE (Run stored procedures)**

**ALL PRIVILEGES (Full administrative access)**

**Principals:** The database user account or collective "role" (a group of users) receiving or losing the permission.

**Deep Dive into DCL Commands**

DCL consists of two fundamental commands: GRANT and REVOKE.

A. The GRANT Command
The GRANT command gives a user or a role specific permissions to interact with database objects (like tables, views, or schemas).

Basic Syntax:

```python
GRANT privilege_name 
ON object_name 
TO user_or_role;
```

Real-World Examples:

Giving Read-Only Access: Allow a data analyst named swapnil_analyst to only query data from the Students table.


```python
GRANT SELECT, INSERT ON Students TO developer_role;
```

# B. The REVOKE Command

The REVOKE command removes previously assigned permissions from a specific user or role, acting as an immediate restriction.

Basic Syntax:

```python
REVOKE privilege_name 
ON object_name 
FROM user_or_role;
```

Real-World Examples:

Removing Destructive Powers: Take away the ability to delete rows from the backend application role.

```python
REVOKE DELETE ON Students FROM developer_role;
```

**Completely Cutting Off Access:** Revoke all administrative powers from a user.

```python
REVOKE ALL PRIVILEGES ON Students FROM swapnil_analyst;
```

***DCL Security Best Practices***

When managing database access control, administrators rely on distinct security strategies to protect data:

The Principle of Least Privilege (PoLP): Users and applications should only be granted the absolute minimum permissions necessary to complete their job. For example, a reporting tool should only receive GRANT SELECT, never GRANT DELETE.

Role-Based Access Control (RBAC): Instead of granting permissions to individual users one by one, permissions are granted to a Role (e.g., HR_Team, Dev_Team). Users are then added to that role, making permission management much cleaner.


# 4. Introduction to DQL (Data Query Language)

Data Query Language (DQL) is the component of SQL dedicated exclusively to retrieving and viewing data from a database. While DDL builds the structure and DML modifies the data, DQL is how you extract meaningful information from your tables.

Core Characteristics of DQL

**Read-Only Operations**: DQL statements do not alter data, change table structures, or consume transactional logs like DML or DDL commands do. They simply read the data in its current state.

**The SELECT Statement**: SELECT is the primary command of DQL.

**Result Sets**: When you execute a DQL query, the database engine returns a temporary visual grid known as a Result Set, which acts like a virtual table matching your criteria.

**Structural Breakdown of a DQL Query**

To fetch exactly what you need, DQL relies on an organized set of clauses. Here is the foundational syntax framework:

```python
SELECT column1, column2 
FROM table_name
WHERE condition
GROUP BY column_name
HAVING group_condition
ORDER BY column_name ASC|DESC
LIMIT number;
```

**1. The SELECT & FROM Clauses (The Pillars)**

SELECT specifies which columns or attributes you want to view.

FROM specifies which table holds those columns.

```python
-- Selecting all columns using the asterisk (*) wildcard
SELECT * FROM Employees;

-- Selecting specific columns (Better for performance)
SELECT FirstName, Salary FROM Employees;
```

**2. Filtering Rows with WHERE**

The WHERE clause screens data row-by-row before any summaries or groupings happen. It drops rows that do not meet your exact conditions.

Operators: =, !=, >, <, >=, <=, AND, OR, NOT.

IN: Matches values within a defined list (WHERE DepartmentID IN (1, 2, 3)).

BETWEEN: Filters values within an inclusive range (WHERE Salary BETWEEN 30000 AND 50000).

LIKE: Used for pattern matching with wildcards:

% matches zero or more characters.

_ matches a single character.


```python
-- Finds employees whose first name starts with 'S'
SELECT * FROM Employees WHERE FirstName LIKE 'S%';
```

**3. Summarizing Data with GROUP BY & HAVING**

GROUP BY collapses rows sharing the same value into a single summary row. It is used alongside aggregate functions like COUNT(), SUM(), AVG(), MIN(), and MAX().

HAVING acts as a WHERE clause, but specifically for groups. It filters the grouped categories after the aggregation takes place.

```python
-- Finds departments where the average salary is greater than 50,000
SELECT DepartmentID, AVG(Salary) AS AvgSalary
FROM Employees
GROUP BY DepartmentID
HAVING AVG(Salary) > 50000;
```

**4. Sorting and Truncating with ORDER BY & LIMIT**

ORDER BY sorts the final display output in ascending (ASC, default) or descending (DESC) order.

LIMIT (or TOP / FETCH depending on your database engine) restricts the maximum number of rows returned in the result set.

```python
-- Gets the top 3 highest-earning employees
SELECT FirstName, Salary 
FROM Employees 
ORDER BY Salary DESC 
LIMIT 3;
```

**Data Aggregation Functions**

When you query data with DQL, you often want to run calculations across columns rather than reviewing rows individually.

 COUNT() ===>  Returns the number of matching rows.

  SUM()  ===>  Calculates the absolute total of a numeric column.

  AVG()	===>   Calculates the mathematical average.

  MAX() ===>   Locates the highest value in a column.

  MIN()	===>   Locates the lowest value in a column.



**4. Query Execution Order (The Logic Engine)**

One of the most critical concepts to master in DQL is the difference between how a query is written versus how the database engine executes it. Writing a query out of logical order will cause execution errors (such as trying to use an alias in a WHERE clause before the SELECT clause has even generated it).

**The Exact Execution Path:**
**FROM & JOIN:** The database locates the target tables and combines rows if necessary.

**WHERE**: Base rows are filtered instantly to minimize the working dataset.

**GROUP BY**: The remaining data rows are divided into aggregate clusters.

**HAVING**: Clustered summary groups are filtered out based on group metrics.

**SELECT**: The engine processes expressions, applies column aliases, and extracts the final column layout.

**ORDER BY**: The final result grid is sorted based on your sorting rules.

**LIMIT**: The output is truncated to the specified number of rows before rendering onto your screen.

# 5. Introduction to TCL (Transaction Control Language)

Transaction Control Language (TCL) is the sub-language of SQL used to manage and maintain data integrity during modifications. It controls how changes made by DML commands (INSERT, UPDATE, DELETE) are saved or undone.

What is a Transaction?
A transaction is a single logical unit of work that contains one or more DML statements. Think of it like an ATM cash withdrawal:

The machine checks your balance.

It deducts the money from your account.

It dispenses the cash.

If any of these steps fail, the entire process must reset to zero. A transaction guarantees an all-or-nothing outcome.

***Core TCL Commands***

There are three primary commands in TCL: COMMIT, ROLLBACK, and SAVEPOINT.

# A. COMMIT
The COMMIT command permanently saves all the changes made during the current transaction to the database disk. Once a COMMIT is executed, the changes cannot be undone using a rollback.

```python
START TRANSACTION;

UPDATE Accounts 
SET Balance = Balance - 5000 
WHERE AccountID = 101;

COMMIT;
```

# B. ROLLBACK
The ROLLBACK command cancels the current transaction and restores the data to its previous state (the state it was in before the transaction started). It acts as an "Undo" button.

```python
START TRANSACTION;

DELETE FROM Employees; -- Accidentally deleting all records!

ROLLBACK;
```

# C. SAVEPOINT
A SAVEPOINT creates a temporary checkpoint within a transaction. This allows you to roll back a specific portion of a transaction instead of undoing the entire thing.

```python
START TRANSACTION;

INSERT INTO Inventory (ItemID, Name) VALUES (1, 'Laptop');
SAVEPOINT Point1; -- Checkpoint 1 created

INSERT INTO Inventory (ItemID, Name) VALUES (2, 'Mouse');
SAVEPOINT Point2; -- Checkpoint 2 created

INSERT INTO Inventory (ItemID, Name) VALUES (3, 'Invalid Item');

-- Realizing the 3rd insert was a mistake, we roll back only to Point2
ROLLBACK TO Point2; 

COMMIT;
```

***The ACID Properties of Transactions***

TCL exists to enforce the ACID properties, which ensure that database transactions are processed reliably:

Atomicity: The entire transaction succeeds, or it completely fails. There is no partial execution (managed via COMMIT and ROLLBACK).

Consistency: A transaction takes the database from one valid state to another valid state, maintaining all structural rules and constraints.

Isolation: Multiple transactions executing concurrently do not interfere with one another. Changes remain invisible to other users until they are committed.

Durability: Once a transaction is committed, its changes are written to permanent storage and will survive even if a system crash or power failure occurs immediately after.

Databases use data types to understand what kind of information to expect in a column, which ensures data integrity and optimizes storage space.


## 1. Numeric Data Types
Used for storing whole numbers, decimals, and financial data.

Exact Numerics (Integers)

**INT / INTEGER:** The standard integer. Typically uses 4 bytes of storage. Range is roughly $-2$ billion to $+2$ billion. Perfect for primary keys and general counts.

**BIGINT:** Used when values exceed the standard integer range (uses 8 bytes). Ideal for massive datasets, global user IDs, or microsecond timestamps.

**SMALLINT / TINYINT:** Smaller storage options (2 bytes and 1 byte respectively). Used for age, months of the year, or status codes to save space.

**Exact Numerics (Decimals/Money)**
*DECIMAL(p, s) / NUMERIC(p, s):*  Used when absolute precision is required, such as financial transactions and currency.

*$p$* (Precision): The total number of digits (both left and right of the decimal point).

*$s$* (Scale): The number of digits to the right of the decimal point.Example: DECIMAL(5, 2) can store values from -999.99 to 999.99.

**Approximate Numerics (Floating-Point)**

*FLOAT / REAL:* Used for scientific calculations where rounding errors are acceptable. They sacrifice exact precision for massive scale. Never use this for money.

# 2. Character and String Data Types
Used to store text, letters, symbols, and alphanumeric data.

Fixed vs. Variable Length
CHAR(n): Fixed-length character data. If you specify CHAR(10) and insert the word "SQL", it will pad the remaining 7 spaces with blanks.

Best use: Consistent length data like state codes ("NY"), country codes ("USA"), or MD5 hashes.

VARCHAR(n): Variable-length character data. If you specify VARCHAR(100) and insert "SQL", it only uses the space for those 3 characters plus a tiny bit of overhead.

Best use: Names, email addresses, and descriptions where length varies.

Large Text Storage

TEXT / VARCHAR(MAX): Used for long-form content like blog posts, logs, or product reviews. Note that TEXT is legacy in some SQL dialects (like SQL Server, which prefers VARCHAR(MAX)), but widely used in others like PostgreSQL.

# 3. Date and Time Data Types
Crucial for temporal tracking, scheduling, and logging events.

**DATE:** Stores just the calendar date formatted as YYYY-MM-DD. (e.g., 2026-06-13).

**TIME:** Stores just the time of day formatted as HH:MM:SS.

**DATETIME / TIMESTAMP:** Stores both date and time (e.g., 2026-06-13 12:30:00).

**Note:** TIMESTAMP is frequently used for "Last Modified" trackings because many systems automatically convert it to UTC or the server's local timezone.

# 4. Binary Data Types
Used to store data exactly as bytes, bypassing text encoding rules.

**BINARY(n):** Fixed-length binary data.

**VARBINARY(n) / BLOB:** Variable-length binary data. BLOB stands for Binary Large Object.

**Best use:** Storing raw data like images, audio files, PDFs, or encrypted keys directly in the database (though modern best practice often stores files on an external cloud server and saves the URL as a VARCHAR instead).

# 5. Miscellaneous & Advanced Data Types
**BOOLEAN:** Stores true/false values. (Note: SQL Server doesn't have a native Boolean data type; it uses BIT where 1 is true and 0 is false).

**JSON:** Increasingly common in modern SQL (PostgreSQL, MySQL). Allows you to store semi-structured JSON strings but query them using special SQL operators.

**UUID / GUID:** Universally Unique Identifier. A 36-character string used as a globally unique key across distributed systems.

# Common Mistakes to Avoid
 1 Using FLOAT for Currency: Float uses binary floating-point math which introduces tiny rounding errors (e.g., 19.99 might become 19.989999). Always use DECIMAL for money.

2 Overestimating CHAR: Don't use CHAR for fields with variable lengths (like CHAR(255) for names) as it wastes database storage by padding empty spaces.

3 Storing Numbers as Strings: Never store phone numbers, zip codes, or IDs as numeric types if you don't intend to do math on them. A zip code like 01234 will lose its leading zero if stored as an INT. Use VARCHAR instead.

# constraints

. **1. What are SQL Constraints?**

Constraints are rules applied to table columns to limit the type of data that can go into them.

**Purpose**: They ensure the accuracy, reliability, and integrity of the data inside the database. If an action (like inserting or updating data) violates a constraint, the database aborts the operation and returns an error.

**Levels:** Constraints can be column-level (applies only to one column) or table-level (applies to multiple columns together).

**2. Main Types of SQL Constraints**

**A. NOT NULL**

By default, a column can hold NULL (empty) values. The NOT NULL constraint enforces a column to always contain a value.

**Use Case:** Crucial for fields that are absolutely required, such as usernames, email addresses, or transaction dates.

Example:

```python
CREATE TABLE Users (
    UserID INT NOT NULL,
    Username VARCHAR(50) NOT NULL,
    Age INT -- This column CAN be left empty (NULL)
);
```

**B. UNIQUE**

The UNIQUE constraint ensures that all values in a column are completely distinct from one another.

Use Case: Perfect for natural unique identifiers like phone numbers, employee IDs, or email addresses.

Key Difference: Unlike a Primary Key, a table can have multiple UNIQUE columns, and they can accept a NULL value (depending on the database system, usually one NULL is allowed).

Example:

```python
SQL
CREATE TABLE Employees (
    EmployeeID INT NOT NULL,
    Email VARCHAR(100) UNIQUE, -- No two employees can have the same email
    PhoneNumber VARCHAR(15) UNIQUE
);
```

**C. PRIMARY KEY**

A PRIMARY KEY uniquely identifies each record in a table. It is essentially a combination of NOT NULL and UNIQUE.

Rules: * A table can have only one primary key.

It cannot contain NULL values.

Example:

```python
CREATE TABLE Products (
    ProductID INT PRIMARY KEY, -- Uniquely identifies each product
    ProductName VARCHAR(100) NOT NULL
);
```

**D. FOREIGN KEY**

A FOREIGN KEY is a column (or collection of columns) in one table that refers to the PRIMARY KEY in another table. It prevents actions that would destroy links between tables (Referential Integrity).

Use Case: Linking an order to a specific customer, or an employee to a specific department.

Example:

```python
-- Parent Table
CREATE TABLE Customers (
    CustomerID INT PRIMARY KEY,
    CustomerName VARCHAR(100)
);

-- Child Table
CREATE TABLE Orders (
    OrderID INT PRIMARY KEY,
    OrderDate DATE,
    CustomerID INT,
    -- Establishes the relationship:
    FOREIGN KEY (CustomerID) REFERENCES Customers(CustomerID)
);
```

**Note:** If you try to insert an order with a CustomerID of 99, but CustomerID 99 does not exist in the Customers table, SQL will reject it.

**E. CHECK**

The CHECK constraint ensures that all values in a column satisfy a specific logical condition (a Boolean expression).

**Use Case:** Restricting ages to positive numbers, ensuring prices are above zero, or limiting status codes to specific words.

Example:

```python
CREATE TABLE Accounts (
    AccountID INT PRIMARY KEY,
    AccountHolder VARCHAR(100),
    Balance DECIMAL(10,2) CHECK (Balance >= 0), -- Balance cannot be negative
    Status VARCHAR(20) CHECK (Status IN ('Active', 'Suspended', 'Closed'))
);
```

**F. DEFAULT**

The DEFAULT constraint provides a fallback value for a column if no value is specified when inserting a new row.

**Use Case:** Setting a default registration date to the current date, or setting an initial account status to "Active".

Example:

```python
CREATE TABLE BlogPosts (
    PostID INT PRIMARY KEY,
    Title VARCHAR(200) NOT NULL,
    Status VARCHAR(20) DEFAULT 'Draft', -- Defaults to 'Draft' if not provided
    CreatedAt TIMESTAMP DEFAULT CURRENT_TIMESTAMP -- Defaults to the current time
);
```

3. Managing Constraints on Existing Tables
You don't just have to add constraints when creating a table; you can alter an existing table to add or remove them.

 *Adding a Constraint*

```python
-- Adding a UNIQUE constraint to an existing column
ALTER TABLE Employees
ADD CONSTRAINT UC_EmployeeEmail UNIQUE (Email);

-- Adding a CHECK constraint to an existing column
ALTER TABLE Employees
ADD CONSTRAINT CHK_EmployeeAge CHECK (Age >= 18);
```

*Dropping (Removing) a Constraint*

```python
-- Syntax varies slightly by database, but generally:
ALTER TABLE Employees
DROP CONSTRAINT CHK_EmployeeAge;
```

**Q1: What is the difference between CHAR and VARCHAR? When would you use one over the other?**

The Answer: *CHAR* is a fixed-length data type. If you define a column as CHAR(10) and insert the word "SQL" (3 characters), it will pad the remaining 7 spaces with blanks, still consuming 10 bytes of storage.

*VARCHAR* is a variable-length data type. If you define it as VARCHAR(10) and insert "SQL", it will only use 3 bytes of storage (plus 1–2 bytes of overhead to record the string length).

*When to use which*: Use CHAR for data where every entry is exactly the same length (e.g., State codes like 'NY', 'CA', or MD5 hashes). Use VARCHAR for data of unpredictable lengths (e.g., names, email addresses).


**Q2: Why should you never use FLOAT or REAL for financial data/currency? What should you use instead?**

**The Answer:**  FLOAT and REAL are approximate numeric data types. They use floating-point math, which can introduce tiny rounding errors (e.g., 19.99 might be stored internally as 19.9899999). While acceptable in scientific calculations, it is disastrous for accounting.

The Solution: Always use DECIMAL(p, s) or NUMERIC(p, s) for currency because they are exact numeric data types that preserve absolute precision.

**Q3: If a phone number or a zip code consists only of numbers, why should you store it as a VARCHAR instead of an INT?**

*The Answer:* There are three primary reasons:

*Leading Zeros*: Integers drop leading zeros. A zip code like 01234 would erroneously save as 1234 in an integer column.

*Mathematical Operations:* You do not perform arithmetic operations (addition, averaging) on phone numbers or zip codes.

*Formatting & Length:* Phone numbers often require characters like +, -, or ( ), which integers cannot store.

# 2. Questions on SQL Constraints
Q4: What is the difference between a PRIMARY KEY and a UNIQUE constraint?

**The Answer:** While both ensure that all values in a column are distinct, they have distinct differences:
| Feature | Primary Key | Unique Key |
|----------|-------------|------------|
| Count per Table | Exactly one primary key allowed per table. | Multiple unique keys allowed per table. |
| NULL Values | Cannot accept `NULL` values. | Can accept `NULL` values (usually one, depending on the RDBMS). |
| Purpose | Used to uniquely identify a row in a table. | Used to prevent duplicate data in non-key columns (e.g., email addresses). |

**Q5: What is a Composite Primary Key, and when would you use it?**

**The Answer:** A Composite Primary Key is a primary key that consists of two or more columns combined together to uniquely identify a row.

*Scenario Example*: Think of a student enrollment table (Student_ID, Course_ID, Enrollment_Date). A student can take multiple courses, and a course can have multiple students. Neither column is unique on its own. However, combining (Student_ID, Course_ID) ensures a student cannot accidentally be enrolled in the exact same class twice.

**Q6: Explain Referential Integrity and how Foreign Keys enforce it. What are ON DELETE CASCADE and ON DELETE SET NULL?**

**The Answer**: Referential Integrity ensures that relationships between tables remain consistent. A FOREIGN KEY enforces this by ensuring you cannot insert a row into a child table with a pointer to a parent row that doesn't exist.

*Handling Deletions:* If you try to delete a row in the parent table that is still referenced by rows in a child table, SQL will block the deletion by default. You can change this behavior with:

*ON DELETE CASCADE:* If the parent row is deleted, all matching child rows are automatically deleted as well.

*ON DELETE SET NULL:* If the parent row is deleted, the child rows are kept, but their foreign key columns are set to NULL.

# 3. Scenario-Based & Practical Interview Questions

**Q7: Look at this query execution order. Why does the WHERE clause throw an error if you try to filter by an aggregate alias?**

Scenario: An interviewer shows you this broken query:

SQL

```python
SELECT department, AVG(salary) AS avg_sal 
FROM employees 
WHERE avg_sal > 50000 -- ERROR HERE!
GROUP BY department;
```

**The Answer:** This fails because of the SQL Order of Execution.
The WHERE clause is executed before the SELECT and GROUP BY clauses. At the moment WHERE filters the rows, the database hasn't grouped the data or calculated the average yet, so it has no idea what avg_sal is.

*The Fix:* To filter based on aggregated results, you must use the HAVING clause, which executes after grouping:

```python
SELECT department, AVG(salary) AS avg_sal 
FROM employees 
GROUP BY department
HAVING AVG(salary) > 50000;
```

**Q8: Can a FOREIGN KEY table column point to a UNIQUE column instead of a PRIMARY KEY column in the parent table?**

**The Answer:** Yes. A foreign key does not strictly have to point to a primary key; it can reference any column in the parent table that has a UNIQUE constraint. The only requirement is that the target column must contain guaranteed unique values so the database can map the relationship unambiguously.

*Numerical order (Lowest–Highest or Highest–Lowest)*

*Date order (Oldest–Newest or Newest–Oldest)*

**What is GROUP BY?**

The GROUP BY clause groups rows that have the same value in one or more columns.

After grouping, aggregate functions are applied to each group.

```python
--Syntax
SELECT column_name, aggregate_function(column_name)
FROM table_name
GROUP BY column_name;

--Example 1: Count Employees in Each Department
SELECT department,
       COUNT(*) AS total_employees
FROM employees
GROUP BY department;
```

output:
| department | total_employees |
| ---------- | --------------- |
| IT         | 3               |
| HR         | 2               |
| Finance    | 1               |


```python
--Example 2: Total Salary Department-wise
SELECT department,
       SUM(salary) AS total_salary
FROM employees
GROUP BY department;
```


output
| department | total_salary |
| ---------- | ------------ |
| IT         | 170000       |
| HR         | 115000       |
| Finance    | 60000        |


```python
--Example 3: Average Salary Department-wise
SELECT department,
       AVG(salary) AS average_salary
FROM employees
GROUP BY department;

--Example 4: Highest Salary Department-wise
SELECT department,
       MAX(salary) AS highest_salary
FROM employees
GROUP BY department;

--Example 5: Lowest Salary Department-wise
SELECT department,
       MIN(salary) AS lowest_salary
FROM employees
GROUP BY department;
```

**GROUP BY Multiple Columns**

SQL can group data by more than one column.

Example:

```python
SELECT department,
       city,
       COUNT(*) AS employees
FROM employees
GROUP BY department, city;
```

output:
| department | city   | employees |
| ---------- | ------ | --------- |
| IT         | Pune   | 2         |
| IT         | Delhi  | 1         |
| HR         | Mumbai | 2         |
| Finance    | Pune   | 1         |


# GROUP BY with WHERE

WHERE filters rows before grouping.

Example:

```python
SELECT department,
       SUM(salary)
FROM employees
WHERE salary > 50000
GROUP BY department;
--Only employees with salary greater than 50000 are grouped.
```

# What is HAVING?

The HAVING clause filters groups, whereas WHERE filters rows.

```python
--Syntax
SELECT column_name,
       aggregate_function(column_name)
FROM table_name
GROUP BY column_name
HAVING condition;

--Example 1: Departments Having More Than 2 Employees
SELECT department,
       COUNT(*) AS total_employees
FROM employees
GROUP BY department
HAVING COUNT(*) > 2;
```

output:
| department | total_employees |
| ---------- | --------------- |
| IT         | 3               |


```python
--Example 2: Departments with Total Salary Greater Than 100000
SELECT department,
       SUM(salary) AS total_salary
FROM employees
GROUP BY department
HAVING SUM(salary) > 100000;
```

output:
| department | total_salary |
| ---------- | ------------ |
| IT         | 170000       |
| HR         | 115000       |


```python
--Example 3: Departments with Average Salary Above 55000
SELECT department,
       AVG(salary) AS avg_salary
FROM employees
GROUP BY department
HAVING AVG(salary) > 55000;
```

**WHERE vs HAVING**
| WHERE                          | HAVING                      |
| ------------------------------ | --------------------------- |
| Filters rows                   | Filters groups              |
| Used before GROUP BY           | Used after GROUP BY         |
| Cannot use aggregate functions | Can use aggregate functions |


**WHERE Example**

```python

SELECT *
FROM employees
WHERE salary > 50000;

--This filters individual employee records.


```

**HAVING Example**

```python
SELECT department,
       AVG(salary)
FROM employees
GROUP BY department
HAVING AVG(salary) > 50000;
--This filters departments based on their average salary.
```

**WHERE + GROUP BY + HAVING**

All three clauses can be used together.

```python
SELECT department,
       COUNT(*) AS total_employees
FROM employees
WHERE salary > 40000
GROUP BY department
HAVING COUNT(*) >= 2;
```

Execution Steps

1:- WHERE filters rows.

2:- GROUP BY creates groups.

3:- Aggregate function is calculated.

4:- HAVING filters groups.

5:- SELECT displays the result.

**SQL Execution Order**

Although we write:

```python
SELECT department,
       COUNT(*)
FROM employees
WHERE salary > 40000
GROUP BY department
HAVING COUNT(*) > 1
ORDER BY department;
```

**SQL processes it in this order:**

FROM

WHERE

GROUP BY

Aggregate Functions

HAVING

SELECT

ORDER BY

Understanding this order helps explain why aggregate functions cannot normally be used in the WHERE clause.

**Common Mistakes**

**Mistake 1: Forgetting GROUP BY**

❌ Wrong

```python
SELECT department,
       COUNT(*)
FROM employees;

--This causes an error because department is neither grouped nor aggregated.

--✔ Correct

SELECT department,
       COUNT(*)
FROM employees
GROUP BY department;
```

**Mistake 2: Using Aggregate Functions in WHERE**

❌ Wrong


```python
SELECT department,
       COUNT(*)
FROM employees
WHERE COUNT(*) > 2
GROUP BY department;

--✔ Correct

SELECT department,
       COUNT(*)
FROM employees
GROUP BY department
HAVING COUNT(*) > 2;
```

# GROUP BY with ORDER BY

Sort departments by highest total salary.

# Interview Questions

**1. What is GROUP BY?**

*Answer*: GROUP BY groups rows having the same values and allows aggregate functions to calculate results for each group.

**2. What is HAVING?**

*Answer*: HAVING filters groups after GROUP BY, usually using aggregate functions.

**3. Difference Between WHERE and HAVING?**

*Answer:*
| WHERE                  | HAVING                      |
| ---------------------- | --------------------------- |
| Filters rows           | Filters groups              |
| Before GROUP BY        | After GROUP BY              |
| No aggregate functions | Aggregate functions allowed |


**4. Can GROUP BY be used without Aggregate Functions?**,

*Answer:* Yes. It returns unique combinations of grouped columns, although DISTINCT is often a simpler choice.

```python
--Example:

SELECT department
FROM employees
GROUP BY department;
```

**5. Can HAVING be used without GROUP BY?**

*Answer*: Yes, in many database systems it can filter the result of an aggregate query without explicit grouping.

```python
--Example:

SELECT COUNT(*) AS total_employees
FROM employees
HAVING COUNT(*) > 5;
```

**Key Points to Remember**

GROUP BY is always used when you want aggregate results per category (department, city, product, etc.).

WHERE filters rows before grouping.

HAVING filters groups after grouping.

Every non-aggregated column in the SELECT list must usually appear in the GROUP BY clause.

GROUP BY and HAVING are among the most frequently asked SQL interview topics.
